## Dataset 2D separabile

```
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
N=200
X0=np.random.randn(N,2)+np.array([-2,0])
X1=np.random.randn(N,2)+np.array([+2,0])
X=np.vstack([X0,X1])
y=np.hstack([np.zeros(N), np.ones(N)])
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

## Perceptron learning rule (scratch)

```
w=np.zeros(2); b=0.0; lr=0.1
for epoch in range(20):
 mistakes=0
 for xi,yi in zip(X,y):
  pred=1.0 if (w@xi+b)>=0 else 0.0
  if pred!=yi:
   upd=lr*(yi-pred)
   w+=upd*xi
   b+=upd
   mistakes+=1
 print(epoch,'mistakes',mistakes)
 if mistakes==0: break
print('w,b',w,b)
```

## Decision boundary

```
xx=np.linspace(X[:,0].min()-1,X[:,0].max()+1,100)
yy=-(w[0]*xx+b)/(w[1]+1e-8)
plt.scatter(X[:,0],X[:,1],c=y)
plt.plot(xx,yy)
plt.show()
```

## Variante PyTorch (opzionale)

```
try:
 import torch, torch.nn as nn
 Xt=torch.tensor(X,dtype=torch.float32)
 yt=torch.tensor(y.reshape(-1,1),dtype=torch.float32)
 model=nn.Sequential(nn.Linear(2,1), nn.Sigmoid())
 opt=torch.optim.SGD(model.parameters(), lr=0.1)
 loss_fn=nn.BCELoss()
 for _ in range(200):
  opt.zero_grad(); out=model(Xt); loss=loss_fn(out,yt); loss.backward(); opt.step()
 print('loss', float(loss))
except Exception as e:
 print('Torch non disponibile:', e)
```
