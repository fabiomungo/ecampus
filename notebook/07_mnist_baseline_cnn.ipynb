## Download MNIST (torchvision)

```
try:
 import torch
 from torch.utils.data import DataLoader
 from torchvision import datasets, transforms
 import matplotlib.pyplot as plt
 tfm=transforms.Compose([transforms.ToTensor()])
 train_ds=datasets.MNIST('./data',train=True,download=True,transform=tfm)
 test_ds=datasets.MNIST('./data',train=False,download=True,transform=tfm)
 train_dl=DataLoader(train_ds,batch_size=128,shuffle=True)
 test_dl=DataLoader(test_ds,batch_size=256,shuffle=False)
 x0,y0=next(iter(train_dl))
 print(x0.shape,y0.shape)
 plt.imshow(x0[0,0], cmap='gray'); plt.title(int(y0[0])); plt.axis('off'); plt.show()
except Exception as e:
 print('Torch/torchvision non disponibili:', e)
```

## Baseline MLP (1 epoca demo)

```
try:
 import torch, torch.nn as nn
 device='cuda' if torch.cuda.is_available() else 'cpu'
 mlp=nn.Sequential(nn.Flatten(), nn.Linear(28*28,256), nn.ReLU(), nn.Linear(256,10)).to(device)
 opt=torch.optim.Adam(mlp.parameters(), lr=1e-3)
 loss_fn=nn.CrossEntropyLoss()
 mlp.train()
 for xb,yb in list(train_dl)[:200]:
  xb,yb=xb.to(device), yb.to(device)
  opt.zero_grad(); logits=mlp(xb); loss=loss_fn(logits,yb); loss.backward(); opt.step()
 mlp.eval(); correct=0; total=0
 with torch.no_grad():
  for xb,yb in list(test_dl)[:50]:
   xb,yb=xb.to(device), yb.to(device)
   pred=mlp(xb).argmax(1)
   correct += (pred==yb).sum().item(); total += len(yb)
 print('acc', correct/total)
except Exception as e:
 print('Torch non disponibile:', e)
```
