## XOR dataset

```
import numpy as np
X=np.array([[0,0],[0,1],[1,0],[1,1]],float)
y=np.array([[0],[1],[1],[0]],float)
print(X)
print(y.T)
```

## Modello lineare: non puÃ² risolvere XOR

```
def sigmoid(z): return 1/(1+np.exp(-z))
w=np.random.randn(2,1); b=0.0
for _ in range(2000):
 p=sigmoid(X@w+b)
 grad_w=X.T@(p-y)/len(X)
 grad_b=np.mean(p-y)
 w-=1.0*grad_w; b-=1.0*grad_b
pred=(sigmoid(X@w+b)>0.5).astype(int)
print('pred linear', pred.T)
```

## MLP 2-2-1 con PyTorch

```
try:
 import torch, torch.nn as nn
 Xt=torch.tensor(X,dtype=torch.float32)
 yt=torch.tensor(y,dtype=torch.float32)
 model=nn.Sequential(nn.Linear(2,2), nn.Tanh(), nn.Linear(2,1), nn.Sigmoid())
 opt=torch.optim.Adam(model.parameters(), lr=0.1)
 loss_fn=nn.BCELoss()
 for _ in range(3000):
  opt.zero_grad(); out=model(Xt); loss=loss_fn(out,yt); loss.backward(); opt.step()
 pred=(model(Xt).detach().numpy()>0.5).astype(int)
 print('pred mlp', pred.T)
except Exception as e:
 print('Torch non disponibile:', e)
```
