Le stringhe sono molto versatili e consentono al programma di elaborare informazioni testuali. stringa costante, racchiusa in citazioni, definisce i valori della stringa letterale all'interno del programma. A volte si può desiderare di definire una maggiore quantità di testo letterale all'interno del programma. Questo testo potrebbe consistere in più righe. La tripla citazione permette di avere più righe di testo.


```python
print(
    """Print
Multiple
Lines
"""
)
```

Come molte lingue Python utilizza una sola (') e una doppia (") cita in modo intercambiabile per indicare le costanti delle stringhe letterali. La convenzione generale è che le doppie citazioni dovrebbero racchiudere il testo reale, come parole o frasi. Le singole citazioni dovrebbero racchiudere il testo simbolico, come i codici di errore. Un esempio di codice di errore potrebbe essere 'HTTP404'.

Tuttavia, non c'è differenza tra singole e doppie citazioni in Python, e si può utilizzare qualsiasi cosa ti piace. Il seguente codice fa uso di un'unica citazione.


```python
print("Hello World")
```

Oltre alle stringhe, Python consente ai numeri come costanti letterali nei programmi. Python include il supporto per punti galleggianti, interi, complessi e altri tipi di numeri. Questo corso non farà uso di numeri complessi. A differenza delle stringhe, le citazioni non racchiudono i numeri.

La presenza di un punto decimale differenzia i numeri a punto variabile e interi. Ad esempio, il valore 42 è un intero. Allo stesso modo, 42.5 è un numero di punto galleggiante. Se si desidera avere un numero di punto mobile, senza una parte di frazione, è necessario specificare una frazione zero. Il valore 42.0 è un numero a punto variabile, anche se non ha parte frazionaria. Come esempio, il seguente codice stampa due numeri.


```python
print(42)
print(42.5)
```

Finora, abbiamo visto solo come definire i valori numerici e stringhe letterali. Questi valori letterali sono costanti e non cambiano quando il programma viene eseguito. Variabili permettono al programma di tenere i valori che possono cambiare mentre il programma viene eseguito. Variabili hanno nomi che permettono di fare riferimento ai loro valori. Il seguente codice assegna un valore intero a una variabile denominata "a" e un valore di stringa a una variabile denominata "b".


```python
a = 10
b = "ten"
print(a)
print(b)
```

La caratteristica chiave delle variabili è che possono cambiare. Il seguente codice dimostra come modificare i valori detenuti dalle variabili.


```python
a = 10
print(a)
a = a + 1
print(a)
```

È possibile mescolare stringhe e variabili per la stampa. Questa tecnica è chiamata una stringa formattata o interpolata. Le variabili devono essere all'interno dei ricci. In Python, questo tipo di stringa è generalmente chiamato f-string. Il f-string è denotato mettendo una "f" proprio di fronte all'apertura singola o doppia citazione che inizia la stringa. Il seguente codice dimostra l'uso di un f-string per mescolare diverse variabili con una stringa letterale.


```python
a = 10
print(f"The value of a is {a}")
```

Puoi anche usare f-strings con matematica (chiamato espressione). Le bretelle ricci possono racchiudere qualsiasi espressione Python valida per la stampa. Il seguente codice dimostra l'uso di un'espressione all'interno dei ricci di un f-string.


```python
a = 10
print(f"The value of a plus 5 is {a+5}")
```

Python ha molti modi per stampare numeri; questi sono tutti corretti. Tuttavia, per questo corso, useremo f-strings. Il seguente codice mostra alcuni dei vari metodi di stampa numeri in Python.


```python
a = 5

print(f"a is {a}")  # Preferred method for this course.
print("a is {}".format(a))
print("a is " + str(a))
print("a is %d" % (a))
```

È possibile utilizzare se-stazioni per eseguire la logica. Avviso i rientri? Questi se-stazioni sono come Python definisce blocchi di codice da eseguire insieme. Un blocco di solito inizia dopo un colon e include tutte le linee allo stesso livello di indent. A differenza di molti altri linguaggi di programmazione, Python utilizza lo spazio bianco per definire blocchi di codice. Il fatto che lo spazio bianco è significativo al significato del codice programma è una fonte frequente di fastidio per i nuovi programmatori di Python. Tab e spazi sono entrambi utilizzati per definire la portata in un programma Python. Mescolare sia gli spazi che le schede nello stesso programma non è raccomandato.


```python
a = 5
if a > 5:
    print("The variable a is greater than 5.")
else:
    print("The variable a is not greater than 5")
```

Il seguente se-stato ha più livelli. Può essere facile identificare questi livelli in modo improprio, quindi fare attenzione. Questo codice contiene un se-stato nidificato sotto il primo "a=5" se-stato. Solo se un è uguale a 5 sarà il nidificato "b=6" se-stato essere eseguito. Inoltre, nota che il comando "elif" significa "else if".


```python
a = 5
b = 6

if a == 5:
    print("The variable a is 5")
    if b == 6:
        print("The variable b is also 6")
elif a == 6:
    print("The variable a is 6")
```

È anche importante notare che il doppio operatore ("=") è usato per testare l'uguaglianza di due espressioni. Il singolo operatore ("=") è utilizzato solo per assegnare valori alle variabili in Python. Il più grande di (">"), meno di ("_HTML_0__="), meno che o uguale ("±=") tutti eseguono come generalmente sarebbe accettato. La prova per disuguaglianza viene eseguita con l'operatore non uguale ("!=").

È comune nei linguaggi di programmazione a loop su una gamma di numeri. Python lo realizza attraverso l'uso dell'operazione **range**. Qui si può vedere un **per** loop e un **range** operazione che causa il programma di loop tra 1 e 3.


```python
for x in range(1, 3):  # If you ever see xrange, you are in Python 2
    print(x)
# If you ever see print x (no parenthesis), you are in Python 2
```

Questo codice illustra alcune incompatibilità tra Python 2 e Python 3. Prima di Python 3, è stato accettabile lasciare le parentesi fuori di una chiamata funzione _print_. Questo metodo di invocare il comando _print_ non è più consentito in Python 3. Allo stesso modo, era un miglioramento delle prestazioni per utilizzare il comando _xrange_ al posto del comando range a volte. Python 3 ha incorporato tutte le funzionalità del comando _xrange_ Python 2 nel comando normale _range_. Di conseguenza, il programmatore non dovrebbe utilizzare il comando _xrange_ in Python 3. Se si vede uno di questi costrutti utilizzati in codice di esempio, allora si sta guardando un vecchio esempio di epoca Python 2.

Il comando _range_ viene utilizzato in combinazione con i loop per passare oltre una specifica gamma di numeri. I casi, dove è necessario scorrere su intervalli di numeri specifici, sono un po 'soliti. Generalmente, i programmatori utilizzano loop sulle collezioni di elementi, piuttosto che i valori numerici di codifica hard nel codice. Collezioni, così come le operazioni che i loop possono eseguire su di loro, è coperto in seguito in questo modulo.

Il seguente è un ulteriore esempio di una stampa a ciclo di stringhe e numeri.


```python
acc = 0
for x in range(1, 3):
    acc += x
    print(f"Adding {x}, sum so far is {acc}")

print(f"Final sum: {acc}")
```
